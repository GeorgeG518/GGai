```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.rcParams['figure.figsize'] = [20, 5]
plt.figure()
plt.subplot(121)

irisvals=[1, 5, 10, 25, 50, 75, 100, 125, 140, 145, 149 ]
catirisvals=[str(each) for each in irisvals]

iris_m_data = np.zeros((len(irisvals)))
iris_stderr_data = np.zeros((len(irisvals)))
                            
for i, each in enumerate(catirisvals):
    arr=np.loadtxt("./iris/iris_"+each+'.dat')
    arr/=int(each)
    iris_stderr_data[i] = np.std(arr) / np.sqrt(len(arr))
    iris_m_data[i] = np.mean(arr)
    
plt.bar(catirisvals, iris_m_data,yerr=1.96*iris_stderr_data,capsize=3)
plt.tight_layout()
plt.title('Iris data accuracy %')
plt.xlabel("Test Set Size [N]")
plt.ylabel("Mean Accuracy [%]")
plt.savefig("iris.jpg",bbox_inches='tight')
plt.show()


```

```python
%matplotlib inline
cancervals=[1, 5, 10, 25, 50, 75, 90, 100, 104]
catcancersvals=[str(each) for each in cancervals]
cancer_stderr_data = np.zeros((len(catcancersvals)))
cancer_m_data = np.zeros((len(catcancersvals)))

                         
for i, each in enumerate(catcancersvals):
    arr=np.loadtxt("./cancer/cancer_"+each+'.dat')
    arr/=int(each)
    cancer_stderr_data[i] = np.std(arr) / np.sqrt(len(arr))
    cancer_m_data[i] = np.mean(arr)
    
plt.title('Cancer data accuracy %')
plt.xlabel("Test Set Size [N]")
plt.ylabel("Mean Accuracy [%]")
plt.tight_layout()
plt.bar(catcancersvals, cancer_m_data,yerr=1.96*cancer_stderr_data,capsize=3)    
plt.savefig("cancer.jpg",bbox_inches='tight')
plt.show()
```
